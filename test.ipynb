```python
import torch
import torch.nn.functional as F

feature_dim = 8192
bs = 8
convolve = True

x = torch.randn(bs, feature_dim)
weights = torch.randn(1, feature_dim)

x = F.rms_norm(x, normalized_shape=[feature_dim])

if convolve:
    logits = F.conv1d(x.unsqueeze(1), weights.unsqueeze(1), padding=feature_dim-1).squeeze(1)
    logits = F.adaptive_max_pool1d(logits, 1)
else:
    logits = torch.matmul(x, weights.T)
logits.shape
```
